```python
import pandas as pd
import numpy as np
data = [
    {"Name": "Asabeneh", "Country":"Finland","City":"Helsinki"},
    {"Name": "David", "Country":"UK","City":"London"},
    {"Name": "John", "Country":"Sweden","City":"Stockholm"}]
df = pd.DataFrame(data)
print(df)
```

```python
weights = [74, 78, 69]
df['Weight'] = weights
df
```

```python
age = [34,67,45]
df["Age"] = age
df
```

```python
df["Mass"] = df['Weight']/df["Age"]*9
df
```

```python
df["Height"] =df["Weight"]*2
df
```

```python
bmi = []
def bmi_cal():
    for w,h in zip(df['Weight'],df["Height"]):
        b = w+(h*h)
        bmi.append(b)
    return bmi
bmi_cal()
df["BMI"] = bmi
df
```

```python
df["Current Year"] = 2023
df["Birth Year"] = df["Current Year"]- df['Age']
df
```
